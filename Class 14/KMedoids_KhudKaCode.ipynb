```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
X = np.array([[2,6],[3,4],[3,8],[4,7],[6,2],[6,4],[7,3],[7,4],[8,5],[7,6]])
```

```python
def findDistance(medians, data):
    
    tot_dist = 0
    
    for point in data:
        distances = [(abs(point - m)).sum() for m in medians]
        min_distance = min(distances)
        tot_dist += min_distance
    
    return tot_dist
    
    
def fit_function(data, k):
    tot = len(data)
    randomPoints = np.sort(np.asarray(random.sample(range(0,tot), k))) #choosen randomly mean points
    
    medians = []
    for i in range(k):
        currentPoint = randomPoints[i]
        medians.append(data[currentPoint])
    
    
    cluster_points = []
    for i in range(k):
        cluster_points.append([])
        
    for point in data:
        distances = [(abs(point - m)).sum() for m in medians]
        min_distance = min(distances)
        classification = distances.index(min_distance)
        cluster_points[classification].append(point)
    
    
    for i in range(k):
        
        allDistances = []
        all_medians = []
        all_cluster_points = []
        
        currentCluster = cluster_points[i]
        
        for p1 in currentCluster:
            new_medians = medians[:]
            new_medians[i] = p1
            
            new_cluster_points = []
            for j in range(k):
                new_cluster_points.append([])

            for point in data:
                distances = [(abs(point - m)).sum() for m in new_medians]
                min_distance = min(distances)
                classification = distances.index(min_distance)
                new_cluster_points[classification].append(point)
    
            currentDistance = findDistance(new_cluster_points, new_medians)
            
            allDistances.append(currentDistance)
            all_medians.append(new_medians)
            all_cluster_points.append(new_cluster_points)
            
        tot_min_distance = min(allDistances)
        index = allDistances.index(tot_min_distance)
        
        medians = all_medians[index]
        cluster_points = all_cluster_points[index]
            
    print(medians)
    print(cluster_points)
    
    # Plotting cluster points
    
    all_points = []
    c = []
    for i in range(len(cluster_points)):
        points = cluster_points[i]
        for point in points:
           # print(point[0])
            all_points.append([point[0],point[1]])
            c.append(i)
    
    all_points = np.array(all_points)
    print(all_points)
    plt.scatter(all_points[:,0],all_points[:,1],c=c)
    plt.show()

fit_function(X, 2)
```
